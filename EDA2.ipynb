```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t=np.arange(0,2,0.1)
t
```

```python
s=1+np.sin(2*np.pi*t)
s
```

```python
fig,ax=plt.subplots()
ax.plot(t,s)
ax.set(xlabel='Time',ylabel='Voltage',title='Time vs Voltage')
plt.grid()
plt.show()
```

```python
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))
```

```python
x=np.linspace(0,7,22)
y=np.linspace(0,15,40)

x,y=np.meshgrid(x,y)
X=x*np.sin(y)
Y=x*np.cos(y)
Z=f(x,y)

fig=plt.figure(figsize=(10,6))
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='magma')
plt.show()
```

```python
ax.view_init(360,0)
fig
```
