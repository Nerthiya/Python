```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv(r"C:\Users\User\Downloads\archive\student_clustering.csv")
df
```

```python
df.plot(x='cgpa',y='iq',kind='scatter',marker='*',color='red')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Student')
plt.figure(figsize=(16,12))
plt.plot
```

```python
df.plot(x='cgpa',y='iq',kind='line')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.groupby('cgpa')['iq'].mean().plot(kind='bar',color='coral')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Student')
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.tight_layout()
plt.show()
```
